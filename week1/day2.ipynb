## Job Summarizer Generator Using Local Model (OLLAMA)

### This is a Job Summarizer Generator which scrapes job contents form website then returns key requirements using local model
#### Model used: Gemma3:1b

```python
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import display, Markdown
from openai import OpenAI
import ollama
```

```python
load_dotenv(override=True)
api_key = os.getenv("OLLAMA_KEY")

if not api_key:
    print("OLLAMA API KEY Not Found. Please check your environment variable")
elif not api_key.startswith("ollama"):
    print("Looks like API KEY does not starts with ' ollama'. Please check if you used the correct API KEY")
else:
    print("API KEY Found and ready for use!!")
```

```python
base_url = os.getenv("BASE_URL")

if not base_url:
    print("Oups!! Base URL Not Found. Please check your environment variable.")
else:
    print("Great! Base URL Found and ready.")
```

```python
system_prompt = """
You are an information extraction engine.

Extract ONLY information explicitly present in the job post.
Do NOT guess or hallucinate missing details.

Return in this format:
- Job Title:
- Experience Required:
- Skills:
- Responsibilities:
- Salary (if mentioned, otherwise "Not specified")
"""
```

```python
user_prompt = "Here is the job post or description. Provide job title, the requirements in bullet points, expirience, salary"
```

```python
def message_prompt(url):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + url}
    ]
    
```

```python
# ollama = OpenAI(base_url=base_url, api_key=api_key)
```

```python
def job_summarize(website_url):
    site_url = fetch_website_contents(website_url)
    response = ollama.chat(model="gemma3:1b", messages=message_prompt(site_url))
    return response["message"]["content"]
```

```python
def job_summary_display(url):
    summary = job_summarize(url)
    display(Markdown(summary))
```

```python
job_summary_display("https://jobs.ea.com/en_US/careers/JobDetail/Senior-SRE-I/211515")
```
