## Email Title Generator
### A summerization application that reads the content of an email and suggest a title

```python
# import modules needed

import os
from dotenv import load_dotenv
from IPython.display import display, Markdown
from google import genai
```

```python
# Check API Key 

load_dotenv(override=True)
api_key = os.getenv("GEMINI_API_KEY")

if not api_key:
    print("API Key not found. Please check your environment variable")
elif not api_key.startswith("AQ."):
    print("API Key found but does not start with 'AQ.'. Please check if you used the correct one")
elif api_key.strip() != api_key:
    print("Looks like your API Key contains spaces. Please remove the trailing whitespaces")
else:
    print("API KEY Found and ready to be used")
```

```python
client = genai.Client(api_key=api_key)
```

```python
system_prompt = "You are a professional assistant. Provide 3 email subject or title based on the email body. dont explain just suggest 3 email. You can start with here are some ..."
```

```python
user_prompt_prefix = "Here is the email body, provide 3 professional email subject or title for this"
```

```python
def both_prompts(email_body):
    return [
        system_prompt,
        user_prompt_prefix + email_body
    ]
```

```python
def suggest_title(email_body):
    email_body = both_prompts(email_body)
    reponse = client.models.generate_content(model="gemini-2.5-flash", contents=email_body)
    return reponse.text
```

```python
def display_title(email_body):
    title = suggest_title(email_body)
    display(Markdown(title))
```
