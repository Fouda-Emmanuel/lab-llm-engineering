## Job Summarizer Generator
### Job summarizer generator that scrapes jobs from websites, summarizes it and generate key requirements of the job

```python
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import display, Markdown
from google import genai
```

```python
load_dotenv(override=True)

api_key = os.getenv("GEMINI_API_KEY")

if not api_key:
    print("No API KEY Found. Please check your environment variable")
elif not api_key.startswith("AQ."):
    print("API Key found but does not starts with 'AQ.'. Please check if you used the correct one")
elif api_key.strip() != api_key:
    print("Looks like your API KEY contains whitespaces. Please remove the trailing whitespaces")
else:
    print("API KEY Found and ready to be used!")
```

```python
client = genai.Client(api_key=api_key)
```

```python
system_prompt = "You are a professional assistant. You are being provided a job description or post and you have to retrieve key requirements of the jobs. Not company name, not mission, not benifits etc just the key requirments on the position"
```

```python
user_prompt = "Here is the job post or description. Provide job title, the requirements in bullet points, expirience, salary"
```

```python
def both_prompts(website):
    return [
        system_prompt,
        user_prompt + website
    ]
```

```python
def job_summarize(url):
    site = fetch_website_contents(url)
    response = client.models.generate_content(model="gemini-2.5-flash", contents=both_prompts(site))
    return response.text
```

```python
def job_summary_display(url):
    summary = job_summarize(url)
    display(Markdown(summary))
```

```python
job_summary_display("https://jobs.ea.com/en_US/careers/JobDetail/Senior-SRE-I/211515")
```
