```python
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
from google import genai
```

```python
load_dotenv(override=True)
api_key = os.getenv("GEMINI_API_KEY")

if not api_key:
    print("No API Key Found. Please check your environment variable")
elif not api_key.startswith("AQ."):
    print("An API Key was found but it does not starts with 'AQ.'. Please check if you are using the correct API Key")
elif api_key.strip() != api_key:
    print("Looks like your API Key contain spaces. Check your environment file again")
else:
    print("API Key Found and ready for use")
```

```python
message= "Wassup Gemini, You Good? This is to test my API Key"
contents = [message]
print("Content to send:", contents)
```

```python
client = genai.Client(api_key=api_key)
response = client.models.generate_content(model="gemini-2.5-flash", contents=contents)
print(response.text)
```

```python
system_prompt = """
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
```

```python
user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.
"""
```

```python
contents = [
    "You are a helpful assistant",
    "Who is Young Thug briefly"
]

response = client.models.generate_content(model="gemini-2.5-flash", contents=contents)
response.text
```

```python
def contents_for(website):
    return [
        system_prompt,
        user_prompt_prefix + website
    ]
```

```python
contents_for(site)
```

```python
def summerize(url):
    website = fetch_website_contents(url)
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=contents_for(website)
        )
    return response.text
```

```python
summerize("https://landlordnde24.com/en-cm/")
```

```python
def display_summary(url):
    summary = summerize(url)
    display(Markdown(summary))
```

```python
display_summary("https://falla237.onrender.com/about")
```
